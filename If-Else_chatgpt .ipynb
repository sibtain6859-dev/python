```python
# Write a program that takes a number as input and prints whether it is even or odd using if-else.

number = int(input('Enter a number : '))
if number % 2 == 0:
    print('Even number')
else:
    print('Odd number')
```

```python
# Take a number from the user and check whether it is positive, negative, or zero.

number = int(input('Enter a number : '))
if number > 0:
    print('Positive')
elif number == 0:
    print('Zero')
else:
    print('Negative')
```

```python
# Write a program to check whether a year is a leap year or not.

year = int(input('Enter a year : '))

if year % 4 == 0 and year % 100 != 0:
    print('Leap year')
elif year % 400 == 0:
    print('Leap year')
else:
    print('Not a leap year')
```

```python
# Ask the user for marks (0–100) and print grades:

# 90–100 → A
# 75–89 → B
# 50–74 → C
# below 50 → Fail

percentage = int(input('Enter marks : '))

if 90 <= percentage <= 100:
    print('Grade A')
elif 75 <= percentage <= 89:
    print('Grade B')
elif 50 <= percentage <= 74:
    print('Grade C')
else:
    print('FAIL')

```

```python
# Take 3 sides as input and check whether they can form a valid triangle or not.

a = int(input('Enter 1st side :'))
b = int(input('Enter 2nd side :'))
c = int(input('Enter 3rd side :'))

if (a + b > c) and (a + c > b) and (b + c > a):
    print('Valid triangle')
else:
    print('NOT valid')
```

```python
# Take two numbers and an operator (+, -, *, /). Perform the operation using if-else conditions.

a = int(input('Enter 1st number :'))
b = int(input('Enter 2nd number :'))
operator = input('Enter operator :')

if operator =='+':
    print(a+b)
elif operator =='-':
    print(a-b)
elif operator =='*':
    print(a*b)
elif operator =='/':
    print(a/b)
else:
    print('Enter the right operator')
```

```python
# Calculate bill based on units:

# 0–100 → ₹5 per unit
# 101–200 → ₹7 per unit
# 201+ → ₹10 per unit
# Use if-else ladder.

unit = int(input('Enter unit :'))

if 0 <= unit <= 100:
    print(5 * unit)
elif 101 <= unit <= 200:
    print(7 * unit)
else:
    print(10 * unit)
    
```
